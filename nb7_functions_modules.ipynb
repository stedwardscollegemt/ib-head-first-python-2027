#### Functions

A function is a "block of code", a sub-program if you like, which is labelled. To run the code inside your function, you simply use the label to invoke it. Naturally, this helps avoiding writing duplicated code.

As one would expect from a sub-program, a function can accept input for processing and can give back a result if needed. 

Before we give you an example, you should keep in mind that functions have a particular *signature*. A signature **defines** the **label** (or name) of the function and its **parameters** (or inputs) if any. The code block must be indented under the signature.

Syntax example:

```python
# define a function
def hello_world():
    print("Hello, world!")

# invoke using the label
hello_world()
```

#### Functions with Parameters and Return Values

Functions with parameters and return values are considered the most useful to learn because they embody the core principles of efficient, modern software development: reusability, modularity, and abstraction. While basic functions can execute a static block of code, functions with parameters and return values turn code into flexible or more generalised "solutions" that accept different inputs, process them, and deliver a tailored result.

Do you remember this code?

```python
# code that reads the file and gives lines...

times = []

for convert in lines:
    minutes = "0"
    rest = convert
    if ":" in convert:
        minutes, rest = convert.split(":")
    seconds, hundredths = rest.split(".")
    converted = (int(minutes) * 60 * 100) + (int(seconds) * 100) + int((hundredths))
    times.append(converted)

print(times)
```

Well, it can look much better if we transfer the string parsing and mathematical conversion in a function:

```python
def convert_str_time_to_hundredths(str_time):
    minutes = "0"
    rest = str_time
    if ":" in str_time:
        minutes, rest = str_time.split(":")
    seconds, hundredths = rest.split(".")
    hundredths_total = (int(minutes) * 60 * 100) + (int(seconds) * 100) + int((hundredths))
    return hundredths_total

# code that reads the file and gives lines...

times = []
for convert in lines:
    converted = convert_str_time_to_hundredths(convert)
    times.append(converted)

print(times)
```

In the original code, the logic for converting time strings into hundredths of a second was written directly inside a loop. This made the code longer, harder to read, and less reusable.

So we moved the conversion logic into a function:

- The function `convert_str_time_to_hundredths(str_time)` takes a time string as a **parameter**.
- It performs all the parsing and calculations internally.
- It **returns** the converted value.

This improves the code in several ways:

- The function can now be used anywhere in the program without rewriting the logic.
- The loop becomes simpler and easier to understand.
- The program is split into smaller, focused parts (one function for conversion, one loop for processing data).
- The details of how the conversion works are "hidden" inside the function.

#### Revisiting Local and Global Variable Scope

We already touched on *variable scope* in Notebook 2:

  > *Variable scope* describes *where* in the program a variable can be used after it has been created...

When we start using functions, understanding scope becomes much more important. As you are already aware, variables generally fall into two main categories: **local** and **global**. 

One can think of scope as "layers in an onion" to better visualise how **variables**, **functions**, and **objects** are nested and accessed within a program in Python.

- **Outermost Layer (Global Scope):** The top layer of the onion represents the global scope. Variables defined here are visible and accessible throughout the entire program — like a bright light shining on the whole onion.
- **Middle Layers (Function/Module Scope):** As you peel back layers, you encounter function scope. Variables declared inside a function are only accessible within that specific function, protecting them from the global space.
- **Innermost Layers (Block Scope):** The tightest, inner layers represent block scope (e.g., inside if statements or for loops). These are highly restricted areas.

Let us analyse our `convert_str_time_to_hundreths` example:
```python
def convert_str_time_to_hundreths(str_time):
    minutes = "0"          # local variable
    rest = str_time        # local variable
    ...
    return hundredths_total
```
Since `minutes`, `rest`, `seconds`, and `hundredths_total` are created inside the function, then they only exist while the function is running. These variables are local and cannot be accessed outside the function.
```python
times = []   # global variable

for convert in lines:
    converted = convert_str_time_to_hundreths(convert)
    times.append(converted)
```
Variables created outside the function. e.g., `times`, can be accessed from most parts of the program. In Python, this can be used *inside* and *outside* our function (though one should be careful of unrestricted access as it can result in unpredictable behaviour in our program).

#### Modules — Always a good idea!

As Paul Barry explains, "you can give anyone a *copy* of your notebook within *their own* Jupyter environment". But, for your Internal Assessment you must build an application that is not restricted to notebooks!

Learning to write simple modules is one of those skills that seems minor at first — but it quietly upgrades everything you do as a programmer. 

We list some reasons that justify using modules here:

**1. You stop writing messy, repetitive code**

When you organize code into modules, you naturally separate concerns. Instead of one giant file doing everything, you break logic into smaller, reusable pieces.

That means:
- Less duplication
- Easier debugging
- Clean and testable structure

**2. Your code becomes reusable**

A module is basically a portable piece of functionality. Once you write it, you can reuse it across projects.

Example:
- A file parser
- A generic random id generator
- An encryption/decryption utility

Write it once but use it everywhere.

**3. You think like a software architect**

Even simple modules force you to ask:

*“What should this piece of code do?”*
*“What inputs and outputs should it have?”*

You are not just coding, but you are structuring logic.

**4. Debugging gets way easier**

If something breaks in a modular system:

- You know exactly where to look
- You can test modules independently

Compare that to digging through a 500-line script 😅.

**5. It prepares you for real-world development**

In professional environments or when making use of popular frameworks:

- Codebases are modular
- Teams work on separate components
- Libraries and packages are just modules

If you understand modules early, you’ll feel much more comfortable learning other modular systems.

**6. It makes collaboration smoother**

Modules create clear boundaries whilst abstracting out complexity:

*“This module handles authentication”*
*“This one handles API calls”*

This means that other people can use your module without needing to understand your entire codebase.

**7. It scales with your projects**

What starts as a small script can grow. If it’s already modular:

- Adding features is easier
- Refactoring is safer
- Maintenance is simpler

Think of it this way:

> A module is like a tool in a toolbox: small, focused, reusable and reliable.

The better your tools, the faster and cleaner you can build things.

#### Activity — Create a statistics module

Let's create a Python module called `stats_utils.py` that includes functions to calculate the following for a list of numbers:

- Mean (average)
- Median
- Mode
- Range
- Standard deviation
- Frequency distribution

Then, import your module inside this notebook and test each function using a sample list of numbers.

For convenience, we asked ChatGPT to give us the Python code for each function. The code makes use of a standard module called `math`. 

Yes, that is right, a module can use other modules!

```python
import math

def mean(data):
    """Return the average of a list of numbers."""
    if len(data) == 0:
        return None
    
    total = sum(data)
    return total / len(data)


def median(data):
    """Return the median (middle value) of a list of numbers."""
    if len(data) == 0:
        return None
    
    sorted_data = sorted(data)
    n = len(sorted_data)
    mid = n // 2

    if n % 2 == 0:
        # Even number of values → average of two middle numbers
        return (sorted_data[mid - 1] + sorted_data[mid]) / 2
    else:
        # Odd number of values → middle value
        return sorted_data[mid]


def mode(data):
    """Return the most frequent value(s) in a list."""
    if len(data) == 0:
        return None
    
    frequency = {}
    
    for value in data:
        if value in frequency:
            frequency[value] += 1
        else:
            frequency[value] = 1

    max_count = max(frequency.values())

    modes = []
    for key, count in frequency.items():
        if count == max_count:
            modes.append(key)

    return modes


def data_range(data):
    """Return the range (max - min) of a list of numbers."""
    if len(data) == 0:
        return None
    
    return max(data) - min(data)


def standard_deviation(data):
    """Return the standard deviation of a list of numbers."""
    if len(data) == 0:
        return None
    
    avg = mean(data)
    
    squared_diffs = []
    for value in data:
        diff = value - avg
        squared_diffs.append(diff ** 2)

    variance = sum(squared_diffs) / len(data)
    return math.sqrt(variance)


def frequency_distribution(data):
    """Return a dictionary counting how often each value appears."""
    frequency = {}
    
    for value in data:
        if value in frequency:
            frequency[value] += 1
        else:
            frequency[value] = 1

    return frequency
```
